```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.array([1, 2, 3, 4, 6, 9, 10, 15, 13])
Y = np.array([2, 6, 8, 11, 15, 21, 30, 40, 34])
```

```python
def plotData(x, y):
    plt.figure(figsize= (6, 3))
    plt.plot(x, y, "ro", ms= 10, mec= "k")
    plt.xlabel("X")
    plt.ylabel("Y")
```

```python
def abline(slope, intercept, epoch):
    plt.figure(figsize= (6, 3))
    axes = plt.gca()
    x_vals = np.array(axes.set_xlim(-20, 20))
    y_vals = intercept + slope * x_vals
    plt.ylim((-50, 50))
    plt.figure(epoch)
    plt.plot(x_vals, y_vals, "--")
    plt.text(0, -4, "epoch " + str(epoch + 1))
```

```python
plotData(X, Y)
```

```python
theta_0 = 1
theta_1 = 0
```

```python
m = 9
lr = 0.025
epochs = 25
```

```python
for epoch in range(epochs):
  sum_0 = 0
  sum_1 = 0
  for i in range(m):
    sum_0 += (theta_0 + theta_1 * X[i] - Y[i])
    sum_1 += (theta_0 + theta_1 * X[i] - Y[i]) * X[i]

  theta_0 = theta_0 - (lr / m) * (sum_0)
  theta_1 = theta_1 - (lr / m) * (sum_1)

  plotData(X, Y)
  abline(theta_1, theta_0, epoch) 
```
