```python
import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import svm
from sklearn.svm import SVC
```

```python
data = loadmat("project7data1.mat")
```

```python
X = data["X"]
y = data["y"]

m = y.shape[0]
```

```python
plt.plot(X[pos, 0], X[pos, 1], "X", mew= 1, mfc= "b", ms= 10, mec= "y")
plt.plot(X[neg, 0], X[neg, 1], "o", mew= 1, mfc= "orange", ms= 10, mec= "r")
plt.grid(True)
```

```python
svm = svm.SVC(kernel= "linear", C= 100)
```

```python
svm.fit(X, y[:,0])
```

```python
w = svm.coef_[0]           
b = svm.intercept_[0]     
x_points = np.linspace(0, 5)    
y_points = -(w[0] / w[1]) * x_points - b / w[1]


plt.plot(X[pos, 0], X[pos, 1], "X", mew= 1, ms= 10, mec= "b")
plt.plot(X[neg, 0], X[neg, 1], "o", mew= 1, mfc= "c", ms= 10, mec= "m")


plt.plot(x_points, y_points, c= "r")
```

```python
m_1 = y_1.shape[0]
```

```python
X_1
```

```python
m_1
```

```python
pos_1 = y_1[:, 0] == 1
neg_1 = y_1[:, 0] == 0
```

```python
pos_1
```

```python
plt.plot(X_1[pos_1, 0], X_1[pos_1, 1], "X", mew= 1, mfc= "orange", ms= 10, mec= "r")
plt.plot(X_1[neg_1, 0], X_1[neg_1, 1], "o", mew= 1, mfc= "m", ms= 10, mec= "b")
plt.grid(True)
```

```python
svm_1 = SVC(kernel= "rbf", C= 1000, gamma= 100)
svm_1.fit(X_1, y_1[:, 0])
```

```python
h= 0.02

x_min, x_max = X_1[:, 0].min() , X_1[:, 0].max()
y_min, y_max = X_1[:, 1].min() , X_1[:, 1].max()
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

Z = svm_1.predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap= plt.cm.coolwarm, alpha= 0.8)

plt.scatter(X_1[:, 0], X_1[:, 1], c= y_1, cmap= plt.cm.coolwarm)

plt.xlabel("Sepal length")
plt.ylabel("Sepal width")

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
```

```python
gamma_selection= [0.1, 1, 10, 100, 1000]
C_selection = [1, 10, 100, 1000]
accuracy= []

for gamma in gamma_selection:
  for c in C_selection:
    svm_2 = SVC(kernel= "rbf", C= c, gamma= gamma)
    svm_2.fit(X_2, y_2[:, 0])
    acc= svm_2.score(X_val, y_val)
    accuracy.append(acc)
    print("accuracy of model for C=%d and gamma=%f is %f" %(c, gamma, acc))
```

```python
svm_2 = SVC(kernel= "rbf", C= 1, gamma= 100)
svm_2.fit(X_2, y_2[:,0])
```

```python
h= 0.02

x_min, x_max = X_2[:, 0].min() , X_2[:, 0].max()
y_min, y_max = X_2[:, 1].min() , X_2[:, 1].max()
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

Z_2 = svm_2.predict(np.c_[xx.ravel(), yy.ravel()])

Z_2 = Z_2.reshape(xx.shape)
plt.contourf(xx, yy, Z_2, cmap= plt.cm.coolwarm, alpha= 0.7)

plt.scatter(X_2[:, 0], X_2[:, 1], c= y_2, cmap= plt.cm.coolwarm)

plt.xlabel("Sepal length")
plt.ylabel("Sepal width")

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
```
